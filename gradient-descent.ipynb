```python
import torch
import matplotlib.pyplot as plt
```

```python
alpha = 0.1
x = torch.tensor([0., 0.], requires_grad=True)
```

```python
function_values = []
for step in range(100):
    function = (1/(1 + torch.exp(-x))).sum()
    print(f'step {step}: {function.item()}')
    function_values.append(function.item())
    function.backward()
    x.data -= alpha * x.grad
    x.grad.zero_()
```

```python
plt.plot(function_values)
plt.xlabel('step')
plt.ylabel('sigmoid($x_1$) + sigmoid($x_2$)')
```
