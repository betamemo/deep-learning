   $$

   $$

### Backward

1. **Gradient of $y$ with respect to $y$**:
   $$
   \frac{\partial y}{\partial y} = 1
   $$

2. **Gradients at $z_8$ and $z_9$**:
   $$
   \frac{\partial y}{\partial z_8} = 1, \quad \frac{\partial y}{\partial z_9} = 1
   $$

3. **Gradients at $z_7$ and $c_1$**:
   $$
   \frac{\partial y}{\partial z_7} = \frac{\partial y}{\partial z_8} \cdot \frac{\partial z_8}{\partial z_7} = 1 \cdot (-3) = -3

   $$
   \frac{\partial y}{\partial c_1} = \frac{\partial y}{\partial z_8} \cdot \frac{\partial z_8}{\partial c_1} = 1 \cdot 3.655 = 3.655
   $$

   $$
   \frac{\partial y}{\partial z_1} = \frac{\partial y}{\partial z_7} \cdot \frac{\partial z_7}{\partial z_1} = -3 \cdot 0.731 = -2.193
   $$

5. **Gradients at $z_4$ and $c_2$**:
   $$
   \frac{\partial y}{\partial z_3} = \frac{\partial y}{\partial z_9} \cdot \frac{\partial z_9}{\partial z_3} + \frac{\partial y}{\partial z_7} \cdot \frac{\partial z_7}{\partial z_4} = 1 \cdot 0.3115 + (-3) \cdot 5 = 0.3115 - 15 = -14.6885

   $$
